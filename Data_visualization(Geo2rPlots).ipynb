```r
#GSE118407_TPM; GRHL2
library(ggplot2)

# Define the data manually
control_values <- c(0.01146944119, 0.01146944119, 0.01146944119)  # Control group
grhl2_values <- c(0.005746118338, 0.03975035467, 0.03975035467, 0.01717014871, 0.01146944119)  # GRHL2 group

# Compute mean and standard deviation
control_mean <- mean(control_values)
control_sd <- sd(control_values)
grhl2_mean <- mean(grhl2_values)
grhl2_sd <- sd(grhl2_values)

# Create a dataframe for plotting
plot_data <- data.frame(
  Condition = c("Control", "shGRHL2"),
  Expression = c(control_mean, grhl2_mean),
  SD = c(control_sd, grhl2_sd)
)

# Plot the graph
ggplot(plot_data, aes(x = Condition, y = Expression , fill = Condition)) +
  geom_bar(stat = "identity", color = "black", width = 1, position = position_dodge(preserve = "single")) +
  geom_errorbar(aes(ymin = Expression - SD, ymax = Expression + SD), width = 0.2) +
  labs(title = "HOTAIR Expression: Control vs. shGRHL2", y = "HOTAIR Expression Level", x = "Condition") +
  theme_minimal() +
  scale_fill_manual(values = c("cyan", "violet")) +
  theme(
    text = element_text(size = 18, face = "bold"),
    axis.title = element_text(size = 18, face = "bold"),
    axis.text = element_text(size = 16, face = "bold"),
    axis.title.y = element_text(margin = margin(r = 20)),
    plot.title = element_text(size = 18, face = "bold", hjust = 0.3, margin = margin(b = 20))
  )
```
